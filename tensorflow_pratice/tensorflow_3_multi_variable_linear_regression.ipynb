```python
import tensorflow as tf
import numpy as np

tf.enable_eager_execution()
tf.set_random_seed(0)
```

```python
x1_data = [1, 0, 3, 0, 5]
x2_data = [0, 2, 0, 4, 0]
y_data  = [1, 2, 3, 4, 5]

learing_rate = tf.Variable(0.001)
#x_data = [x1_data,x2_data]

w1 = tf.Variable(tf.random_uniform([1],-10.0, 10.0)) #1차원 : vector 의 형태로 uniform 분포에서 random값 가져옴
w2 = tf.Variable(tf.random_uniform([1],-10.0, 10.0))

for i in range(1001):
    with tf.GradientTape() as tape:
        h = w1 * x1_data + w2 * x2_data
        cost = tf.reduce_mean(tf.square(h - y_data))
    w1_grad, w2_grad = tape.gradient(cost,[w1,w2])
    
    w1.assign_sub(learing_rate * w1_grad)
    w2.assign_sub(learing_rate * w2_grad)
    
    if i % 50 == 0:
        print("{:5} | {:10.6f} | {:10.4f} | {:10.4f}".format(
          i, cost.numpy(), w1.numpy()[0], w2.numpy()[0]))
```

```python
x_data = [
    [1., 0., 3., 0., 5.],
    [0., 2., 0., 4., 0.]
]
w = tf.Variable(tf.random_uniform([1,2],-1.0,1.0))
b = tf.Variable(tf.random_uniform([1],-1.0,1.0))

for i in range(1001):
    with tf.GradientTape() as tape:
        h = tf.matmul(w,x_data) + b # [1,2] * [2.5] = [1,5]
        cost = tf.reduce_mean(tf.square(tf.subtract(h,y_data)))
    
        w_grad,b_grad = tape.gradient(cost,[w,b])
        w.assign_sub(learing_rate * w_grad)
        b.assign_sub(learing_rate * b_grad)
    
    if i % 50 == 0:
        print("{:5} | {:10.6f} | {:10.4f} | {:10.4f} | {:10.6f}".format(
            i, cost.numpy(), w.numpy()[0][0], w.numpy()[0][1], b.numpy()[0]))
        
```
